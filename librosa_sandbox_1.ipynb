```python
import matplotlib.pyplot as plot
import matplotlib.style as plot_style
from IPython.display import Audio
import librosa
import numpy as np
import librosa.display
```

```python
filename = librosa.util.example_audio_file()               # audio sample
y, sr = librosa.load(filename, offset=25.0, duration=20.0) # just a fragment
spectogram = np.abs(librosa.stft(y))                       # short time fourier transform

chroma = librosa.feature.chroma_cqt(y=y,sr=sr)
tonnetz = librosa.feature.tonnetz(y=y, sr=sr)
```

```python
librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
plot.show()
```

```python
librosa.display.specshow(tonnetz, y_axis='tonnetz', x_axis='time')
plot.show()
```
